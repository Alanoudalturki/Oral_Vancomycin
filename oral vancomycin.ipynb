```python
# ==============================================
# CDI – One-Stop Publication Script (Tables+Figs)
# ==============================================
# Requires: pandas, numpy, seaborn, matplotlib, statsmodels, lifelines, openpyxl
# pip install pandas numpy seaborn matplotlib statsmodels lifelines openpyxl

import os, warnings, itertools
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import statsmodels.formula.api as smf
from scipy.stats import kruskal, chi2_contingency, fisher_exact

from lifelines import KaplanMeierFitter, CoxPHFitter
from lifelines.statistics import logrank_test

warnings.filterwarnings("ignore", message="Data Validation extension is not supported")

# --------------------------------
# 0) Paths & output directories
# --------------------------------
DATA_PATH = 'from nbformat import v4 as nbf
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer
from reportlab.lib.styles import getSampleStyleSheet
import os

# Create a dummy notebook structure with code cells and markdown outputs
nb = nbf.new_notebook()
nb.cells = [
    nbf.new_markdown_cell("# Oral Vancomycin CDI Study\n\nComprehensive analysis, tables, figures, and results."),
    nbf.new_code_cell("## Code sections would go here (data loading, cleaning, stats, visualization)"),
    nbf.new_markdown_cell("## Outputs: Tables & Figures\n- Table 1: Baseline Characteristics\n- Table 2A/B: By Dose\n- Table 3–4: Outcomes\n- Table 5–6: Regression Models\n- Figures 1–8: Distributions, KM, Mortality, Forest Plots"),
    nbf.new_markdown_cell("## Results\nDetailed narrative and statistical findings."),
    nbf.new_markdown_cell("## Discussion\nPhD-level discussion, implications, and recommendations.")
]

# Save notebook as .ipynb
notebook_path = "/mnt/data/oral_vancomycin.ipynb"
import nbformat
with open(notebook_path, "w") as f:
    nbformat.write(nb, f)

# Now create a PDF summary (not full Jupyter execution, but structured text)
pdf_path = "/mnt/data/oral_vancomycin.pdf"
doc = SimpleDocTemplate(pdf_path)
styles = getSampleStyleSheet()
story = []

story.append(Paragraph("Oral Vancomycin CDI Study", styles["Title"]))
story.append(Spacer(1, 12))

sections = [
    "Methods: Study design, measures, and statistical analysis were implemented using Python 3.11 (pandas, numpy, scipy, lifelines, statsmodels).",
    "Results: Tables 1–6 and Figures 1–8 summarize baseline characteristics, outcomes, and multivariable analyses.",
    "Discussion: Findings suggest no significant benefit of escalating oral vancomycin beyond 125 mg for non-fulminant CDI. Implications for stewardship, cost reduction, and future research are discussed."
]
for sec in sections:
    story.append(Paragraph(sec, styles["Normal"]))
    story.append(Spacer(1, 12))

doc.build(story)

notebook_path, pdf_path
' 
OUT_DIR   = "cdi_outputs"
FIG_DIR   = os.path.join(OUT_DIR, "figures")
os.makedirs(FIG_DIR, exist_ok=True)

# --------------------------
# 1) Load & normalize names
# --------------------------
df = pd.read_excel(DATA_PATH)
df.columns = (df.columns
              .str.strip()
              .str.lower()
              .str.replace(r"[^\w]+", "_", regex=True))

# --------------------------
# 2) Flexible column pickers
# --------------------------
def pick_col(must_include=None, any_of=None):
    must_include = [t.lower() for t in (must_include or [])]
    any_of = [t.lower() for t in (any_of or [])]
    def ok(col):
        c = col.lower()
        return all(t in c for t in must_include) and (not any_of or any(t in c for t in any_of))
    cands = [c for c in df.columns if ok(c)]
    if not cands and any_of:
        cands = [c for c in df.columns if all(t in c.lower() for t in must_include)]
    return sorted(cands, key=lambda x: (len(x), x))[0] if cands else None

dose_col = pick_col(["vancomycin"], ["dosage","dose","mg"]) or "vancomycin_dosage__mg_"
ttr_col  = pick_col(["time","symptom","resolution"], ["day","days"])
mort_col = pick_col(["mortality"], ["90","day","days"]) or "90_daysmortality"
rec_col  = pick_col(["recurrence"], ["90","day","days"])
sev_col  = pick_col(["severity"], ["cdi","severe","fulminant","non"])

age_col    = pick_col(["age"])
gender_col = pick_col(["gender"])
wt_col     = pick_col(["weight"], ["kg"])
wbc_col    = pick_col(["wbc"], ["15000"]) or pick_col(["wbc"])
scr_col    = "scr" if "scr" in df.columns else None

# optional clinical flags
hypo_col  = pick_col(["hypotension"]) or pick_col(["shock"])
ileus_col = pick_col(["ileus"])
mega_col  = pick_col(["megacolon"])

diab_col  = pick_col(["diabetes"])
immun_col = pick_col(["immunosuppression"])
renal_col = pick_col(["renal","kidney"])

freq_col  = pick_col(["frequency"])
dur_col   = pick_col(["duration"])
route_col = pick_col(["route"])
comp_col  = pick_col(["compliance"])
adj_col   = pick_col(["adjunct"])
ae_col    = pick_col(["adverse_events"])
chg_col   = pick_col(["vancomycin_change_during"])
chg_dose_col = pick_col(["vancomycin_change_dose"])
abx_during_col = pick_col(["use_of_other_antibiotics_during"])
abx_after_col  = pick_col(["use_of_other_antibiotics_after"])

# --------------------------------
# 3) Derived / cleaned variables
# --------------------------------
yn_map = {"yes":1,"y":1,"1":1,"true":1,"no":0,"n":0,"0":0,"false":0}

# Dose numeric (125/250/500)
df["dose"] = pd.to_numeric(df[dose_col].astype(str).str.extract(r"(\d+)")[0], errors="coerce")

# Time-to-resolution (days)
df["ttr_days"] = pd.to_numeric(df[ttr_col].astype(str).str.extract(r"(\d+)")[0], errors="coerce")

# Binary outcomes
df["mort90"] = df[mort_col].astype(str).str.strip().str.lower().map(yn_map).fillna(0).astype(int) if mort_col else 0
if rec_col is not None:
    df["recurrence"] = df[rec_col].astype(str).str.strip().str.lower().map(yn_map).fillna(0).astype(int)
else:
    df["recurrence"] = 0
if ae_col:
    df["adverse_event"] = df[ae_col].astype(str).str.strip().str.lower().map(yn_map).fillna(0).astype(int)
else:
    df["adverse_event"] = 0

# WBC category — robust (avoid dtype promotion)
def build_cat_from_sign(series_str, gt_label, lt_label):
    out = pd.Series(pd.NA, index=series_str.index, dtype="object")
    out.loc[series_str.str.contains(">", na=False)] = gt_label
    out.loc[series_str.str.contains("<", na=False)] = lt_label
    return out

if wbc_col:
    s = df[wbc_col].astype(str)
    df["wbc_cat"] = build_cat_from_sign(s, ">15000", "<15000")
else:
    df["wbc_cat"] = pd.Series(pd.NA, index=df.index, dtype="object")

# SCr binary
if scr_col:
    s = df[scr_col].astype(str)
    scr_bin = pd.Series(pd.NA, index=df.index, dtype="object")
    scr_bin.loc[s.str.startswith(">", na=False)] = "High (≥1.5)"
    scr_bin.loc[s.str.startswith("<", na=False)] = "Low (<1.5)"
    df["scr_binary"] = scr_bin
else:
    df["scr_binary"] = pd.Series(pd.NA, index=df.index, dtype="object")

# Simple binaries
def bin_clean(s):
    return (s.astype(str).str.strip().str.lower()
            .map({"yes":"Yes","y":"Yes","1":"Yes","true":"Yes","no":"No","n":"No","0":"No","false":"No"}))

if diab_col:  df["diabetes_b"] = bin_clean(df[diab_col]).fillna("No")
if immun_col: df["immuno_b"]   = bin_clean(df[immun_col]).fillna("No")
if renal_col: df["renal_b"]    = bin_clean(df[renal_col]).fillna("No")
if chg_col:   df["dose_changed"] = bin_clean(df[chg_col]).fillna("No")
if comp_col:  df["compliant"]    = bin_clean(df[comp_col]).fillna("Yes")
if abx_during_col: df["abx_during"] = bin_clean(df[abx_during_col]).fillna("No")
if abx_after_col:  df["abx_after"]  = bin_clean(df[abx_after_col]).fillna("No")
if adj_col:   df["adjunct"]     = df[adj_col].astype(str).str.strip()

# Severity: prefer provided; else derive. Always keep as Series.
if sev_col:
    sev_src = df[sev_col].astype(str).str.strip().str.lower()
    sev_std = pd.Series(pd.NA, index=df.index, dtype="object")
    sev_std.loc[sev_src.str.contains("fulm", na=False)] = "Fulminant"
    sev_std.loc[sev_src.str.contains("sev",  na=False)] = "Severe"
    sev_std.loc[sev_src.str.contains("non",  na=False)] = "Non-severe"
else:
    sev_std = pd.Series(pd.NA, index=df.index, dtype="object")

# fill missing from clinical rules
fulm_flag = pd.Series(False, index=df.index)
for cc in [hypo_col, ileus_col, mega_col]:
    if cc and cc in df:
        fulm_flag = fulm_flag | df[cc].astype(str).str.lower().isin(["yes","y","1","true"]).fillna(False)
wbc_high = df["wbc_cat"].eq(">15000").fillna(False)
scr_high = df["scr_binary"].eq("High (≥1.5)").fillna(False)

mask_missing = sev_std.isna()
sev_std.loc[mask_missing] = np.where(
    fulm_flag[mask_missing], "Fulminant",
    np.where(wbc_high[mask_missing] | scr_high[mask_missing], "Severe", "Non-severe")
)
df["severity"] = sev_std.astype("object")

# --------------------------------
# 4) Helper utilities for tables
# --------------------------------
dose_levels = [125.0, 250.0, 500.0]

def fmt_mean_sd(s):
    s = pd.to_numeric(s, errors="coerce").dropna()
    return "—" if s.empty else f"{np.mean(s):.2f} ± {np.std(s, ddof=1):.2f}"

def fmt_med_iqr(s):
    s = pd.to_numeric(s, errors="coerce").dropna()
    if s.empty: return "—"
    q1, q2, q3 = np.percentile(s, [25,50,75])
    return f"{q2:.1f} [{q1:.1f}, {q3:.1f}]"

def percent(n, d):
    return f"{int(n)} ({(100*n/d):.1f}%)" if d and d>0 else "0 (0.0%)"

def kruskal_safe(groups):
    arrs = [pd.to_numeric(g, errors="coerce").dropna().values for g in groups]
    arrs = [a for a in arrs if len(a) > 0]
    if len(arrs) < 2: return np.nan
    allv = np.concatenate(arrs)
    if np.allclose(allv, allv[0]): return np.nan
    try:
        return kruskal(*arrs).pvalue
    except ValueError:
        return np.nan

def chi2_or_fisher(ct):
    t = ct.loc[(ct.sum(1) > 0), (ct.sum(0) > 0)]
    if t.size == 0 or t.shape[0] < 2 or t.shape[1] < 2:
        return (np.nan, "n/a", "insufficient categories")
    chi2, p, dof, exp = chi2_contingency(t, correction=False)
    if t.shape == (2,2) and (exp < 5).any():
        p = fisher_exact(t)[1]; test="fisher"
    else:
        test="chi2"
    note = ">20% expected<5" if (exp < 5).sum()/exp.size > 0.2 else "ok"
    return (p, test, note)

def p_fmt(p): return "—" if pd.isna(p) else ("<0.001" if p < 1e-3 else f"{p:.4f}")

# ------------------------------------------
# 5) Table 1 — Overall baseline
# ------------------------------------------
rows = []

# Continuous overall
for c,name in [(age_col,"Age (years)"), (wt_col,"Weight (kg)"), ("ttr_days","Time to symptom resolution (days)")]:
    if c and c in df.columns:
        s = df[c]
        rows.append([name,
                     f"Mean ± SD: {fmt_mean_sd(s)}; Median [IQR]: {fmt_med_iqr(s)}",
                     f"n={s.notna().sum()}"])

# Key categorical overall
def add_cat_overall(col, nice):
    if col and col in df.columns:
        s = df[col].astype(str).replace({"nan":"Missing"})
        rows.append([nice, "", f"N={s.notna().sum()}"])
        for lvl, n in s.value_counts(dropna=False).items():
            rows.append([f"  - {lvl}", "", percent(n, len(s))])

add_cat_overall(gender_col, "Gender")
add_cat_overall("diabetes_b", "Diabetes")
add_cat_overall("immuno_b", "Immunosuppression")
add_cat_overall("renal_b", "Renal disease")
add_cat_overall("severity", "Severity")
add_cat_overall("wbc_cat", "WBC category")
add_cat_overall("scr_binary", "SCr category")

table1 = pd.DataFrame(rows, columns=["Variable", "Categories / Statistics", "N (%) or Mean ± SD"])

# -------------------------------------------------
# 6) Table 2A/2B — Baseline by dose
# -------------------------------------------------
# 2A continuous by dose
cont_cols = []
for c in [age_col, wt_col, "ttr_days"]:
    if c and c in df.columns: cont_cols.append(c)

t2a_rows=[]
for col in cont_cols:
    label = {"ttr_days":"Ttr Days"}.get(col, ("Age (years)" if col==age_col else ("Weight (kg)" if col==wt_col else col)))
    per_dose=[]
    for d in dose_levels:
        s = df.loc[df["dose"]==d, col]
        per_dose.append(f"{fmt_mean_sd(s)}; {fmt_med_iqr(s)} (n={s.notna().sum()})" if s.notna().sum() else "—")
    overall = f"{fmt_mean_sd(df[col])}; {fmt_med_iqr(df[col])} (n={df[col].notna().sum()})"
    pval = kruskal_safe([df.loc[df["dose"]==d, col] for d in dose_levels])
    t2a_rows.append([label, *per_dose, overall, p_fmt(pval)])
table2A = pd.DataFrame(t2a_rows, columns=["Variable","125 mg","250 mg","500 mg","Overall","P-value"])

# 2B categorical by dose
cat_cols = []
for c,name in [(gender_col,"Gender"),("diabetes_b","Diabetes"),("immuno_b","Immunosuppression"),
               ("renal_b","Renal disease"),("severity","Severity"),
               ("wbc_cat","WBC category"),("scr_binary","SCr category")]:
    if c and c in df.columns: cat_cols.append((c,name))

t2b_rows=[]
for col, nice in cat_cols:
    s = df[col].astype(str)
    ct = pd.crosstab(df["dose"], s).reindex(index=dose_levels, fill_value=0)
    p, test, note = chi2_or_fisher(ct)
    for i,lvl in enumerate(ct.columns):
        r=[nice, lvl]
        for d in dose_levels:
            denom = (df["dose"]==d).sum()
            r.append(percent(ct.loc[d,lvl], denom) if denom else "—")
        r.append(percent(ct[lvl].sum(), len(df)))
        r.append(p_fmt(p) if i==0 else "")
        t2b_rows.append(r)
table2B = pd.DataFrame(t2b_rows, columns=["Variable","Level","125 mg","250 mg","500 mg","Overall","P-value"])

# ------------------------------------------
# 7) Table 3 — Outcomes by dose
# ------------------------------------------
def rate_by_dose(col): return df.groupby("dose")[col].mean().reindex(dose_levels).mul(100)

t3 = pd.DataFrame({
    "MORT90 %": rate_by_dose("mort90"),
    "RECURRENCE %": rate_by_dose("recurrence") if "recurrence" in df else 0,
    "ADVERSE %": rate_by_dose("adverse_event")
}).round(1)
t3.loc["Overall"] = [df["mort90"].mean()*100,
                     (df["recurrence"].mean()*100 if "recurrence" in df else 0),
                     df["adverse_event"].mean()*100]
table3 = t3.reset_index().rename(columns={"index":"Dose"})

# ------------------------------------------
# 8) Table 4 — TTR distribution (% by dose)
# ------------------------------------------
pref_days = [3,7,10,14,28]
ttr_vals = sorted(df["ttr_days"].dropna().unique().astype(int))
order = [d for d in pref_days if d in ttr_vals] + [d for d in ttr_vals if d not in pref_days]
ct_ttr = pd.crosstab(df["ttr_days"].astype("Int64"), df["dose"]).reindex(index=order, columns=dose_levels, fill_value=0)
table4 = (ct_ttr.apply(lambda c: 100*c/c.sum() if c.sum() else c)).round(1).reset_index().rename(columns={"ttr_days":"Resolution day"})

# ------------------------------------------
# 9) Models — Logistic (mortality) & Cox PH
# ------------------------------------------
logit_or_table = None
if df["mort90"].nunique() > 1:
    terms = ["C(dose)"]
    if age_col: terms.append(age_col)
    if gender_col: terms.append(f"C({gender_col})")
    if "diabetes_b" in df: terms.append("C(diabetes_b)")
    if "immuno_b" in df:   terms.append("C(immuno_b)")
    if "renal_b"  in df:   terms.append("C(renal_b)")
    formula = "mort90 ~ " + " + ".join(terms)
    try:
        logit = smf.logit(formula, data=df.dropna(subset=["mort90"])).fit(disp=0)
        params = logit.params; conf = logit.conf_int()
        logit_or_table = pd.DataFrame({
            "Variable": params.index,
            "OR": np.exp(params),
            "CI Lower": np.exp(conf[0]),
            "CI Upper": np.exp(conf[1]),
            "p-value": logit.pvalues
        }).reset_index(drop=True)
    except Exception as e:
        print("Logistic model skipped:", e)

cox_hr_table = None
cph = None
cox_cols = ["ttr_days","dose"] + [c for c in [age_col, gender_col, "diabetes_b","immuno_b","renal_b","severity"] if c and c in df.columns]
cox_df = df[cox_cols].dropna(subset=["ttr_days"]).copy()
if not cox_df.empty and cox_df["ttr_days"].nunique() > 1 and cox_df.shape[0] > 10:
    cat_vars = [v for v in ["gender", "diabetes_b","immuno_b","renal_b","severity"] if v in cox_df.columns]
    cox_df = pd.get_dummies(cox_df, columns=cat_vars, drop_first=True)
    cox_df["event"] = 1
    try:
        cph = CoxPHFitter()
        cph.fit(cox_df, duration_col="ttr_days", event_col="event")
        summ = cph.summary.copy().rename_axis("Variable").reset_index()
        need = ["exp(coef)","exp(coef) lower 95%","exp(coef) upper 95%","p"]
        have = [c for c in need if c in summ.columns]
        cox_hr_table = summ[["Variable", *have]].rename(columns={
            "exp(coef)":"HR","exp(coef) lower 95%":"CI Lower",
            "exp(coef) upper 95%":"CI Upper","p":"p-value"
        })
    except Exception as e:
        print("Cox PH model skipped:", e)

# --------------------------
# 10) Save tables to Excel
# --------------------------
excel_path = os.path.join(OUT_DIR, "CDI_Publication_Tables.xlsx")
with pd.ExcelWriter(excel_path, engine="openpyxl") as writer:
    table1.to_excel(writer, sheet_name="Table 1 (Overall Baseline)", index=False)
    table2A.to_excel(writer, sheet_name="Table 2A (Continuous by dose)", index=False)
    table2B.to_excel(writer, sheet_name="Table 2B (Categorical by dose)", index=False)
    table3.to_excel(writer, sheet_name="Table 3 (Outcomes by dose)", index=False)
    table4.to_excel(writer, sheet_name="Table 4 (TTR % by dose)", index=False)
    if logit_or_table is not None:
        logit_or_table.to_excel(writer, sheet_name="Logit ORs (Mortality)", index=False)
    if cox_hr_table is not None:
        cox_hr_table.to_excel(writer, sheet_name="Cox HRs (TTR)", index=False)
print(f"[saved] {excel_path}")

# =========================
# 11) FIGURES (Blues theme)
# =========================
plt.rcParams.update({
    "figure.dpi": 150,
    "font.size": 12,
    "axes.titlesize": 18,
    "axes.labelsize": 13,
    "axes.titleweight": "bold",
    "axes.labelweight": "bold",
    "axes.grid": True,
    "grid.alpha": 0.25
})
sns.set_style("whitegrid")

BLUE_DARK, BLUE_MED, BLUE_LIGHT, BLUE_P1, BLUE_P2 = "#08306B","#2171B5","#6BAED6","#9ECAE1","#C6DBEF"
BLUES3 = [BLUE_DARK, BLUE_MED, BLUE_LIGHT]
BLUES5 = [BLUE_DARK, BLUE_MED, BLUE_LIGHT, BLUE_P1, BLUE_P2]
RED_REF = "#E41A1C"

def save_show(fig, name):
    path = os.path.join(FIG_DIR, name)
    fig.tight_layout()
    fig.savefig(path, bbox_inches="tight")
    print(f"[saved] {path}")
    plt.show()

# FIG A — KM by dose with 50% red line & dose medians
kmf = KaplanMeierFitter()
fig, ax = plt.subplots(figsize=(8,5))
for i, d in enumerate(sorted(df["dose"].dropna().unique())):
    m = (df["dose"]==d) & df["ttr_days"].notna()
    if m.sum() == 0: continue
    kmf.fit(df.loc[m,"ttr_days"], event_observed=np.ones(m.sum()), label=f"{int(d)} mg")
    kmf.plot_survival_function(ax=ax, ci_show=False, color=BLUES3[i%3], lw=2.5)
    med = kmf.median_survival_time_
    if pd.notna(med):
        ax.axvline(med, color=BLUES3[i%3], linestyle=":", lw=1.7, alpha=0.9)
        ax.text(med, 0.53, f" {int(d)}mg med≈{med:.0f}d", color=BLUES3[i%3], va="bottom", ha="left", fontsize=10)
ax.axhline(0.5, color=RED_REF, linestyle="--", lw=1.5, alpha=0.9)
ax.set_title("Kaplan–Meier: Symptom Persistence by Vancomycin Dose")
ax.set_xlabel("Days"); ax.set_ylabel("Proportion Symptomatic"); ax.set_ylim(0,1.02)
save_show(fig, "figA_km_by_dose.png")

# FIG B — Time-to-resolution stacked % by dose
pref_days = [3,7,10,14,28]
ttr_vals = sorted(df["ttr_days"].dropna().unique().astype(int))
order = [d for d in pref_days if d in ttr_vals] + [d for d in ttr_vals if d not in pref_days]
ct_pct = (pd.crosstab(df["ttr_days"].astype("Int64"), df["dose"])
            .reindex(index=order, fill_value=0))
ct_pct = ct_pct.apply(lambda c: 100*c/c.sum() if c.sum() else c)
fig, ax = plt.subplots(figsize=(8,5))
bottom = np.zeros(len(ct_pct.columns), dtype=float)
xlabels = [f"{int(c)} mg" for c in ct_pct.columns]
for i, day in enumerate(ct_pct.index):
    vals = ct_pct.loc[day].values.astype(float)
    ax.bar(xlabels, vals, bottom=bottom, color=BLUES5[i%5], edgecolor="white", linewidth=0.4, label=str(int(day)))
    bottom += vals
ax.set_title("Time to Resolution by Dose (Stacked %)"); ax.set_xlabel("Dose"); ax.set_ylabel("% of patients")
ax.legend(title="Resolution day", frameon=True, edgecolor="none")
save_show(fig, "figB_ttr_stacked_by_dose.png")

# FIG C — Mortality by dose + overall reference line
m = df.groupby("dose")["mort90"].mean().reindex(dose_levels)*100
fig, ax = plt.subplots(figsize=(7,5))
ax.bar([f"{int(x)} mg" for x in m.index], m.values, color=[BLUE_DARK, BLUE_MED, BLUE_LIGHT])
overall = df["mort90"].mean()*100
ax.axhline(overall, color=RED_REF, linestyle="--", lw=1.5, alpha=0.9)
ax.text(0.02, overall+1, f"Overall: {overall:.1f}%", color=RED_REF, fontsize=10)
ax.set_title("90-day Mortality by Dose"); ax.set_ylabel("% of patients"); ax.set_xlabel("Dose")
save_show(fig, "figC_mortality_by_dose.png")

# FIG D — Adverse events by dose
a = df.groupby("dose")["adverse_event"].mean().reindex(dose_levels)*100
fig, ax = plt.subplots(figsize=(7,5))
ax.bar([f"{int(x)} mg" for x in a.index], a.values, color=[BLUE_DARK, BLUE_MED, BLUE_LIGHT])
ax.set_title("Adverse Events by Dose"); ax.set_ylabel("% of patients"); ax.set_xlabel("Dose")
save_show(fig, "figD_adverse_by_dose.png")

# FIG E — Distributions (Age, Weight, TTR) with median red line
def hist_kde(series, title, xlabel, fname):
    s = pd.to_numeric(series, errors="coerce").dropna()
    if s.empty: return
    fig, ax = plt.subplots(figsize=(7.5,4.8))
    sns.histplot(s, bins=20, stat="count", kde=True, color=BLUE_MED, edgecolor="white", linewidth=0.3, ax=ax)
    med = s.median()
    ax.axvline(med, color=RED_REF, linestyle="--", lw=1.5, alpha=0.9)
    ax.set_title(title); ax.set_xlabel(xlabel); ax.set_ylabel("Count")
    ax.text(med, ax.get_ylim()[1]*0.9, f" median={med:.1f}", color=RED_REF, ha="left", va="top", fontsize=10)
    save_show(fig, fname)

if age_col: hist_kde(df[age_col], "Distribution: Age (years)", "Age (years)", "figE1_hist_age.png")
if wt_col:  hist_kde(df[wt_col],  "Distribution: Weight (kg)", "Weight (kg)", "figE2_hist_weight.png")
hist_kde(df["ttr_days"], "Distribution: Time to symptom resolution (days)", "Time to symptom resolution (days)", "figE3_hist_ttr.png")

# FIG F — Logistic OR forest (if model fit)
if 'logit' in locals() and logit_or_table is not None and not logit_or_table.empty:
    or_df = logit_or_table.copy()
    or_df = or_df[or_df["Variable"]!="Intercept"]
    fig, ax = plt.subplots(figsize=(7, max(3, 0.4*len(or_df))))
    y = np.arange(len(or_df))
    ax.errorbar(or_df["OR"], y, xerr=[or_df["OR"]-or_df["CI Lower"], or_df["CI Upper"]-or_df["OR"]],
                fmt="o", color=BLUE_MED, ecolor=BLUE_P1, capsize=3, lw=1.5)
    ax.axvline(1.0, color=RED_REF, linestyle="--", lw=1.5, alpha=0.9)
    ax.set_yticks(y); ax.set_yticklabels(or_df["Variable"])
    ax.set_xscale("log")
    ax.set_title("Logistic Regression: Mortality (Odds Ratios)")
    ax.set_xlabel("Odds Ratio (log scale)")
    save_show(fig, "figF_logit_or_forest.png")

# FIG G — Cox HR forest (if model fit)
if cph is not None and cox_hr_table is not None and not cox_hr_table.empty:
    hr_df = cox_hr_table.copy()
    if {"HR","CI Lower","CI Upper"}.issubset(hr_df.columns):
        fig, ax = plt.subplots(figsize=(7, max(3, 0.4*len(hr_df))))
        y = np.arange(len(hr_df))
        ax.errorbar(hr_df["HR"], y, xerr=[hr_df["HR"]-hr_df["CI Lower"], hr_df["CI Upper"]-hr_df["HR"]],
                    fmt="o", color=BLUE_MED, ecolor=BLUE_P1, capsize=3, lw=1.5)
        ax.axvline(1.0, color=RED_REF, linestyle="--", lw=1.5, alpha=0.9)
        ax.set_yticks(y); ax.set_yticklabels(hr_df["Variable"])
        ax.set_xscale("log")
        ax.set_title("Cox PH: Time to Resolution (Hazard Ratios)")
        ax.set_xlabel("Hazard Ratio (log scale)")
        save_show(fig, "figG_cox_hr_forest.png")

# -------------------------------
# 12) Pairwise log-rank by severity
# -------------------------------
if "severity" in df.columns and df["severity"].notna().any():
    print("\n=== Pairwise Log-Rank Tests by Severity Strata ===")
    for sev in df["severity"].dropna().unique():
        sub = df[df["severity"]==sev]
        doses = sorted(sub["dose"].dropna().unique())
        if len(doses) < 2: 
            continue
        print(f"\nSeverity: {sev}")
        for d1, d2 in itertools.combinations(doses, 2):
            g1 = sub[sub["dose"]==d1]["ttr_days"].dropna()
            g2 = sub[sub["dose"]==d2]["ttr_days"].dropna()
            if len(g1)>0 and len(g2)>0:
                res = logrank_test(g1, g2, event_observed_A=np.ones(len(g1)), event_observed_B=np.ones(len(g2)))
                print(f"  {int(d1)} mg vs {int(d2)} mg → p = {res.p_value:.4f}")

# -------------------------------
# 13) Console print tables (nice)
# -------------------------------
def print_table(title, df_):
    print("\n" + "="*86)
    print(title)
    print("="*86)
    print(df_.to_string(index=False))

print_table(f"Table 1. Baseline Characteristics of Patients with CDI (N = {len(df)})", table1)
print_table("Table 2A. Continuous Variables by Vancomycin Dose", table2A)
print_table("Table 2B. Categorical Variables by Vancomycin Dose", table2B)
print_table("Table 3. Outcomes by Vancomycin Dose", table3)
print_table("Table 4. Distribution of Time-to-Resolution by Dose (% of patients)", table4)

if logit_or_table is not None:
    print_table("Logistic Regression (Mortality): Odds Ratios", logit_or_table.round(3))
if cox_hr_table is not None:
    print_table("Cox PH (Time to Resolution): Hazard Ratios", cox_hr_table.round(3))
```
